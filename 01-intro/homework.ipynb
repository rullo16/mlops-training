```python
import pandas as pd
import sklearn
```

```python
df_january = pd.read_parquet(
    "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet",
    engine="pyarrow",
)
df_february = pd.read_parquet(
    "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-02.parquet",
    engine="pyarrow"
)
df = pd.concat((df_january, df_february))
```

```python
df.head()
```

```python
df.describe()
```

```python
print("Columns: ",df.shape[-1])
```

```python
pickup_time = pd.to_datetime(df.tpep_pickup_datetime)
dropoff_time = pd.to_datetime(df.tpep_dropoff_datetime)
```

```python
df['duration'] = dropoff_time - pickup_time
```

```python
df['duration'] = df.duration.apply(lambda td: td.total_seconds()/60)
```
